# Optuna Dashboard Launcher

Acest notebook permite vizualizarea interactivă a studiilor de optimizare Optuna salvate anterior în bazele de date SQLite.
Rulând celulele de mai jos, veți putea lansa un server web local (dashboard) dedicat fiecărui model. 

### 1. Vizualizare Optimizare XGBoost
Dashboard-ul va fi disponibil pe portul **8090**.

```python
import optuna
from optuna_dashboard import run_server

storage_url_xgb = "sqlite:///outputs/optuna_studies/xgboost_optimization.db"

try:
    storage = optuna.storages.RDBStorage(storage_url_xgb)
    print("Se lansează Optuna Dashboard pentru XGBoost...")
    print("Accesati in browser: http://127.0.0.1:8090")
    run_server(storage, host="127.0.0.1", port=8090)
except Exception as e:
    print(f"Eroare la pornirea dashboard-ului XGBoost: {e}")
```
